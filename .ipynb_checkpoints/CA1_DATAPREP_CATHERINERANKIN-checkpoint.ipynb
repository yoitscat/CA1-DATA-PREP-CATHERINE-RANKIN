In this case, our dependent variable is our 'Target', with all others being independent variables, hence we will drop the Target feature and create a new dataset without this feature, so we can preform PCA

```python
notarget=newdf.drop(columns=['Target'])
```

```python
notarget.head()
```

<font size="5"><b>FIRST PCA</b></font>

```python
pca=PCA().fit(notarget)
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('Number of Components')
plt.ylabel('Cumulative Explained Variance')
```

This graph shows that in order to retain at least 99.5% of the variance/signal in the data, we need to have approximately 2 components and to retain 100% we need at approximately 15 components 

```python
pca = PCA(2)
projected = pca.fit_transform(notarget)
```

We have created 2 new features here needed to retain 99.5% of the signal

```python
projected.shape
```

```python
studentdata_pca = pd.DataFrame(projected, columns =['C1', 'C2'])
```

```python
studentdata_pca.head()
```

Here we have created new values that match the signal in our original dataset

```python
studentdata_pca["Target"] = newdf["Target"]
```

```python
studentdata_pca.head()
```

```python
studentdata_array = studentdata_pca.values
```

```python
studentdata_array
```

I will use a 70/30 split to train the model

```python
from sklearn.model_selection import train_test_split
X = studentdata_array[:,0:2]
y = studentdata_array[:,2]
X_train, X_validation, Y_train, Y_validation = train_test_split(X, y, test_size=0.30, random_state=1)
```

```python
from matplotlib import pyplot
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import StratifiedKFold
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.svm import SVC
```

```python
models = []
models.append(('LR', LogisticRegression(solver='liblinear', multi_class='ovr')))
models.append(('LDA', LinearDiscriminantAnalysis()))
models.append(('KNN', KNeighborsClassifier()))
models.append(('CART', DecisionTreeClassifier()))
models.append(('NB', GaussianNB()))
models.append(('SVM', SVC(gamma='auto')))
```

```python
results = []
names = []
for name, model in models:
	kfold = StratifiedKFold(n_splits=10, random_state=1, shuffle=True)
	cv_results = cross_val_score(model,X_train, Y_train, cv=kfold, scoring='accuracy')
	results.append(cv_results)
	names.append(name)
	print('%s: %f (%f)' % (name, cv_results.mean(), cv_results.std()))
```

The performace of this model based on accuracy levels is poor.

```python
pyplot.boxplot(results,labels=names)
pyplot.title('Algorithm Comparison')
```

```python
model=LogisticRegression(solver='liblinear', multi_class='ovr')
model.fit(X_train, Y_train)
predictions = model.predict(X_validation)
```

```python
print(accuracy_score(Y_validation, predictions))
print(confusion_matrix(Y_validation, predictions))
```

```python
conf_matrix = np.array([[0, 2, 404],
                             [0, 1, 240],
                             [0, 0, 681]])

# Plot the confusion matrix as a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues')
plt.xlabel('Predictions')
plt.ylabel('True')
plt.title('Confusion Matrix')
plt.show()
```

As we can see from this confusion matrix, this model did well at predicting the 3rd class which was graduating, however it preformed poorly with the other 2 classes, dropout and enrolled

<font size='5'><b> Curse of Dimensionality, Model Preformance before and after PCA</b></font>

To compare, I will evaluate the performance of the model using the original data, before PCA is preformed.

```python
studentdata3_array = newdf.values
```

```python
newdf.shape
```

```python
X = studentdata3_array[:,0:36]
y = studentdata3_array[:,36]
X_train, X_validation, Y_train, Y_validation = train_test_split(X, y, test_size=0.30, random_state=1)
```

```python
results = []
names = []
for name, model in models:
	kfold = StratifiedKFold(n_splits=10, random_state=1, shuffle=True)
	cv_results = cross_val_score(model,X_train, Y_train, cv=kfold, scoring='accuracy')
	results.append(cv_results)
	names.append(name)
	print('%s: %f (%f)' % (name, cv_results.mean(), cv_results.std()))
```

```python
model=LogisticRegression(solver='liblinear', multi_class='ovr')
model.fit(X_train, Y_train)
predictions = model.predict(X_validation)
```

```python
print(accuracy_score(Y_validation, predictions))
print(confusion_matrix(Y_validation, predictions))
```

```python
confu_matrix = np.array([[323, 27, 56],
                        [62, 55, 124],
                        [22, 17, 642]])

# Plot the confusion matrix as a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(confu_matrix, annot=True, fmt='d', cmap='Blues')
plt.xlabel('Predictions')
plt.ylabel('True')
plt.title('Confusion Matrix')
plt.show()
```

'The curse of dimensionality' can arise due to sparsity and overfitting. As the number of features within a dataset increases, the amount of data needed to cover the space grows. If there is not enough data to cover the feature space, or too many 0's, this sparsity can make it difficult to predict accurately the relationships within the data. On the other hand, high dimensional datasets run the risk of models overfitting the data, as they may create relationships from the noise in the datase,  relationships which do not accurately generalise the data for future predictions. 

Hence, it is crucial to appropriately match the model complexity with the dataset's dimensionality to achieve optimal performance, as well as conducting EDA and PCA, if necessary.

It appears that the model preforms better on the data before PCA, than it does after PCA. The model may be too simplistic to capture the relationships within the data, the dataset may be too non-linear to capture relationships accurately. Preforming PCA on this dataset may have caused important information to be lost, reducing dimenstionality too far, however high variance among the principal components was still retained after excecuting PCA.
