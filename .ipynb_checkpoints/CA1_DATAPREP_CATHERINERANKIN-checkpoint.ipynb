I have used this piece of code to see if any of my columns still have objects in them, this has highlighted 'Target'

As the feature 'Target' is an object, I will apply a numeric classification to it, similar to other features in the dataset

```python
replacement_dict= {'Dropout': '0', 'Drop': '0', 'dropout':'0', 'Enrolled':'1', 'Graduate':'2', 'Grad':'2', 'graduate':'2'} 
newdf['Target'] = newdf['Target'].replace(replacement_dict)
```

```python
newdf['Target'].unique()
```

```python
corr_matrix = newdf.corr()

# Plot the heatmap
plt.figure(figsize=(50, 48))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Matrix Heatmap')
plt.show()
```

We can see from this heatmap that Age of enrollment, debtor and gender have negative correlations with the target, while modules and grades have stronger positive correlations with the target.
